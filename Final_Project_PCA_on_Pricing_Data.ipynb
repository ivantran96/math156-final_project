```python
#compute daily returns
daily_returns = data['Adj Close'].pct_change(1)
daily_returns = daily_returns.iloc[1:]
daily_returns.dropna(axis=1, inplace=True)
daily_returns
```

```python
def pca_svd(X):
    # center data
    X = X - X.mean(axis=0)
    
    # calculate SVD
    U, s, Vt = np.linalg.svd(X, full_matrices=False)
    
    return Vt.T
```

```python
SPY = pd.read_csv('SPY.csv', index_col=0)

# calcualte daily returns
SPY_daily_ret = SPY['Adj Close'].pct_change(1).iloc[1:]

# calculate cumulative returns
SPY_cumret = (SPY_daily_ret + 1).cumprod()
SPY_cumret.index = pd.to_datetime(SPY_cumret.index)
```

```python
pca = pca_svd(daily_returns)

for i in range(20):
    # use first principle component as weights for each ticker
    pc = pca[:, i]
    weights = (pc - min(pc)) / (max(pc) - min(pc))
    weights = weights / sum(weights)

    # use the weights to make PCA portfolio and calculate cumulative returns
    cumret = (daily_returns+1).cumprod(0) @ weights
    cumret.index = pd.to_datetime(cumret.index)
    
    df = pd.concat([cumret, SPY_cumret], 1)
    df.columns = ["PCA Portfolio", "S&P500"]
    df.plot()
```

```python
#convert daily returns dataframe to matrix
X=daily_returns.to_numpy()
```

```python
#perform SVD using the covariance matrix
mean = np.mean(X, axis=0)
X_centered = X-mean.reshape(1,-1)
S = X_centered.T@X_centered
D, V = np.linalg.eig(S)


order = D.argsort()[::-1]
V = V[:, order]
```

```python
#See how many pcs needed to account for most of the variance
num_pcs=53 #need 53 pcs to explain 75% of variance
sum=0
for i in range(0,num_pcs):
  sum+=D[i]

print(sum/np.sum(D))
```

```python
for i in range(0,num_pcs):
  print(D[i]/np.sum(D))

#first pc explains significantly more than all the others
```

```python
#find indices of top 10 weights in each PC
pcs=[]
for col in range(V.shape[1]):
    v=V[:,col]
    pcs.append(v)

ind = np.argpartition(pcs[0], -10)[-10:]
ind=ind[np.argsort(pcs[0][ind])]
index=[]
for i in range(len(pcs)):
  ind_top = np.argpartition(pcs[i], -10)[-10:]
  ind_top=ind_top[np.argsort(pcs[i][ind_top])]
  index.append(ind_top)
```

```python
#partition the dataframe to include highest weighted stocks only
top_10=[]
for j in range(len(index)):
  top10=daily_returns.iloc[:,index[j]].mean(1)
  top_10.append(top10)
```

```python
#plot portfolio of highest weighted stocks against SPY
for i in range(len(top_10)-450):
  mycrs = top_10[-(i+1)].cumsum().apply(np.exp)
  mycrs.index = pd.to_datetime(mycrs.index)
  t10 = pd.concat([mycrs, SPY_cumret], 1)
  t10.columns = ["Top 10 PCA Portfolio", "S&P500"]
  t10.plot()
```

```python
#compute log returns for vectorized backtesting
daily_logreturns = data['Adj Close'].apply(np.log).diff(1)
daily_logreturns = daily_returns.iloc[1:]
daily_logreturns.dropna(axis=1, inplace=True)
top10log=daily_logreturns.iloc[:,ind]
```

```python
#vectorized backtesting for highest weighted stock in PC1
import matplotlib.pyplot as plt
rs = data['Adj Close']['CZR'].apply(np.log).diff(1)
w1 = 5 # short-term moving average window
w2 = 22 # long-term moving average window
ma_x = data['Adj Close']['CZR'].rolling(w1).mean() - data['Adj Close']['CZR'].rolling(w2).mean()
pos = ma_x.apply(np.sign) # +1 if long, -1 if short
fig, ax = plt.subplots(2,1)
ma_x.index = pd.to_datetime(ma_x.index)
ma_x.plot(ax=ax[0], title='Moving Average Cross-Over')
pos.plot(ax=ax[1], title='Position')
```

```python
#performance plot under trading strategy
my_rs = pos.shift(1)*rs
my_rs.index = pd.to_datetime(my_rs.index)
my_rs.cumsum().apply(np.exp).plot(title='Strategy Performance')
```

```python
#vectorized backtesting for 10 highest weights from PC1
rs = top10log
w1 = 5
w2 = 22
ma_x = top10log.rolling(w1).mean() - top10log.rolling(w2).mean()
pos = ma_x.apply(np.sign)
pos /= pos.abs().sum(1).values.reshape(-1,1)
fig, ax = plt.subplots(2,1)
ma_x.index = pd.to_datetime(ma_x.index)
ma_x.plot(ax=ax[0], title='Moving Average Cross-Overs')
ax[0].legend(bbox_to_anchor=(1.1, 1.05))
pos.plot(ax=ax[1], title='Position')
ax[1].legend(bbox_to_anchor=(1.1, 1.05))
```

```python
my_rs = (pos.shift(1)*rs)
my_rs.index = pd.to_datetime(my_rs.index)
my_rs.cumsum().apply(np.exp).plot(title='Strategy Performance')
```

```python
my_rs2 = (pos.shift(1)*rs).sum(1)
my_rs2.index = pd.to_datetime(my_rs2.index)
t10_vect = pd.concat([my_rs2.cumsum().apply(np.exp), SPY_cumret], 1)
t10_vect.columns = ["Long/Short PCA Portfolio", "S&P500"]
t10_vect.plot()
```

```python
#vectorized backtesting under exponential moving average strategy
rs= top10log
w1 = 5
w2 = 22
max = top10log.ewm(w1).mean() - top10log.ewm(w2).mean()
position = max.apply(np.sign)
position /= position.abs().sum(1).values.reshape(-1,1)
figure, axis = plt.subplots(2,1)
max.index = pd.to_datetime(max.index)
max.plot(ax=axis[0], title='Exp Moving Average Cross-Overs')
axis[0].legend(bbox_to_anchor=(1.1, 1.05))
position.plot(ax=axis[1], title='Position')
axis[1].legend(bbox_to_anchor=(1.1, 1.05))
```

```python
my_ret = (position.shift(1)*rs)
my_ret.index = pd.to_datetime(my_ret.index)
my_ret.cumsum().apply(np.exp).plot(title='EMA Strategy Performance')
```

```python
my_ret = (position.shift(1)*rs).sum(1)
my_ret.index = pd.to_datetime(my_ret.index)
t10_ewm = pd.concat([my_ret.cumsum().apply(np.exp), SPY_cumret], 1)
t10_ewm.columns = ["PCA Portfolio", "S&P500"]
t10_ewm.plot()
```
